```python
import pandas as pd
import seaborn as se
import statistics as st
```

```python
# Importa os dados
tempo = pd.read_excel ("/content/drive/MyDrive/Colab Notebooks/tempo.xlsx")
```

```python

# Visualiza os dados
tempo
tempo.head(20)
tempo.tail(20)
tempo.sample(20)
```

```python
# Verifica o tamanho com as quantidades de linhas e colunas
tempo.shape
```

```python
# Dá nomes às colunas
tempo.columns = ["condição de tempo",
      "temperatura", "umidade", "vento",
      "jogar"]
```

```python
# Visualiza os dados
tempo.head(20)
```

```python
# Agrupa os dados categóricos por condição de tempo
condicao_de_tempo = tempo.groupby(["condição de tempo"]).size()
condicao_de_tempo
```

```python
# Exibe condição de tempo agrupada em um gráfico barra
condicao_de_tempo.plot.bar()
```

```python

# Agrupa a coluna númerica temperatura
'''
Seleciona a coluna 'temperatura' do DataFrame dataset e calcula estatísticas descritivas
[média (mean), desvio padrão (std), mínimo (min), máximo (max), quartis (25%, 75%),
mediana (50%), número de valores não nulos (count)]
para os valores nessa coluna, fornecendo um resumo estatístico da distribuição.
'''
tempo["temperatura"].describe()
```

```python
# Exibe temperatura em um gráfico boxplot com título temperatura
se.boxplot(tempo['temperatura']).set_title('Temperatura')
```

```python
# Exibe a temperatura em um gráfico histograma com título temperatura
se.histplot(tempo["temperatura"]).set_title("Temperatura")
```

```python
# Agrupa a coluna númerica umidade
'''
Seleciona a coluna 'umidade' do DataFrame dataset e calcula estatísticas descritivas
[média (mean), desvio padrão (std), mínimo (min), máximo (max), quartis (25%, 75%),
mediana (50%), número de valores não nulos (count)]
para os valores nessa coluna, fornecendo um resumo estatístico da distribuição.
'''
tempo["umidade"].describe()
```

```python
# Exibe umidade em um gráfico boxplot com título umidade
se.boxplot(tempo['umidade']).set_title('umidade')
```

```python
# Exibe os umidade em um gráfico histograma com título umidade
se.histplot(tempo["umidade"]).set_title("umidade")
```

```python
# Agrupa os dados categóricos por vento
vento = tempo.groupby(["vento"]).size()
vento
```

```python
# Exibe os dados agrupados da variável 'vento' em um gráfico de barras
vento.plot.bar()
```

```python
# Agrupa os dados categóricos por jogar
jogar = tempo.groupby(["jogar"]).size()
jogar
```

```python
# Exibe os dados agrupados da variável 'jogar' em um gráfico de barras
jogar.plot.bar()
```
